# 1. Data Set selection
  * One for regression, one for classification
  * Define what your objectives are.

1. Regression: Medical charges dataset, aiming to predict an approximation of the medical costs of a patient given various factors such as age, sex, bmi, humber of children, smoker, and the section of USA the patient is located.
2. Classification: Diabetes dataset was chosen, with end goal being to predict being positive for diabetes or negative. Logistic regression is a good idea here.



# 2. Data Cleaning
 * handling missing values and outliers.
 * Do any needed transformations

```python
#imports
import pandas as pd
import numpy as np
import scipy 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, BayesianRidge, LinearRegression
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, mean_absolute_error, r2_score, confusion_matrix
from sklearn import tree
```

```python
#load data
datac=pd.read_csv("diabetes.csv")
datar=pd.read_csv("insurance.csv")
```

```python
#clean data - check for null values
nan_count = datac.isnull().sum()
print('Number of NaN (diabetes):', nan_count)

nan_count = datar.isnull().sum()
print('Number of NaN (medical fees):', nan_count)
```

```python
#Check datatypes
print(datac.dtypes)
print(datar.dtypes)
#we will need to convert the object columns into int columns for the purposes of regression
```

```python
datar['sex'] = datar['sex'].map({'female': 1, 'male': 0})
datar['region'] = datar['region'].map({'northwest': 0, 'northeast': 1, 'southeast':2, 'southwest':3})
datar['smoker'] = datar['smoker'].map({'yes': 1, 'no': 0})
datar
```

# 3. Model Implementation:
   * regression: apply 2 regression techniques (linear regression, decision trees, etc)
   * classification: implement two classification mehtods (e.g. logistic regression, SVM)

```python
#implementing models

#REGRESSION
r_features=['age', 'sex', 'bmi', 'children', 'smoker', 'region']
r_target='charges'

X=datar[r_features]
y=datar[r_target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=39)

rsplit=[X_train, X_test, y_train, y_test]
li_model=LinearRegression()
li_model=li_model.fit(X_train,y_train)
br_model=BayesianRidge()
br_model=br_model.fit(X_train,y_train)


#CLASSIFICATION 
# Split the dataset into test and train
c_features=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X=datac[c_features]
y=datac['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
csplit=[X_train, X_test, y_train, y_test]
lr_model = LogisticRegression(max_iter=300)
lr_model.fit(X_train, y_train)  # Train the model

dt_model=tree.DecisionTreeClassifier()
dt_model = dt_model.fit(X_train, y_train)


#tree.plot_tree(dt_model)
```

# 4. Model Evaluation:
* evaluate model with appropriate metrics
* compare performance of models, discuss findings

```python
#evaluate regression models

li_pred=li_model.predict(rsplit[1])
li_mae = mean_absolute_error(rsplit[3], li_pred)
li_r2 = r2_score(rsplit[3], li_pred)

print ("Linear Regression")
print("LR mae:", li_mae)
print("LR R2:", li_r2)

br_pred=br_model.predict(rsplit[1])
br_mae=mean_absolute_error(rsplit[3], br_pred)
br_r2 = r2_score(rsplit[3], br_pred)

print ("Bayesian Ridge Regression")
print("BR mae:", br_mae)
print("BR R2:", br_r2)


#Evaluate classification models:

lr_predictions = lr_model.predict(csplit[1])
lr_acc=accuracy_score(csplit[3], lr_predictions)
lr_ar_score=roc_auc_score(csplit[3], lr_predictions)


print ("Logistic Regression accuracy", lr_acc)
print ("Logistic Regression ROC-AUC-Score", lr_ar_score)
print(classification_report(csplit[3], lr_predictions))

dt_predictions=dt_model.predict(csplit[1])
dt_acc=accuracy_score(csplit[3], dt_predictions)
dt_ar_score=roc_auc_score(csplit[3], dt_predictions)

print ("Data Tree Classifier accuracy", dt_acc)
print ("Data Tree ROC-AUC-Score", dt_ar_score)
print(classification_report(csplit[3], dt_predictions))

```

5. Result Interpretation
* Interpret results of models
*  Create visualizations 

Firstly, comparing our two regression models: Linear Regression has a *very slightly* lower mean-absolute error, while also having a *very slightly* higher R^2 score. A 0.75 isn't terrible on a scale of 0~1, so we can say that both models are at least decent fits. However, the error is somewhat large at $4000 or so on average. Perhaps there are outliers that could be removed...?

Secondly, looking at our classifiers: We have better results all across the board for the Data Tree Classifier. Accuracy, precision, recall, AUC-ROC Score, F1-Score. The Logistic Regression model has okay scores - it's good at predicting the 0 outcome, but suffers at the 1 outcome. The Data Tree excels all across the board - however, we must be careful, as Data Trees tend to be vulnerable to overfitting.

```python
import seaborn as sns
import matplotlib.pyplot as plt


fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()

for i, var in enumerate(['age', 'sex', 'bmi', 'children', 'smoker', 'region']):
    sns.regplot(x=var, y='charges', data=datar, ax=axs[i], scatter_kws={'alpha':0.5}, line_kws={'color': 'blue'})
    axs[i].set_xlabel(var.capitalize()) 
    axs[i].set_ylabel('Charges')  

plt.show()
```

```python
#visualize regression models

#visualize linear regression
plt.figure()
sns.regplot 
plt.show()

#visualize bayesian ridge regression
```

```python
#visualize classifiers 
#visualize logistic regression

conmat = confusion_matrix(csplit[3], lr_predictions)

val = np.mat(conmat) 
classnames = list(set(csplit[2]))
df_cm = pd.DataFrame(
        val, index=classnames, columns=classnames, 
    )

plt.figure()
heatmap = sns.heatmap(df_cm, annot=True, cmap="Reds")
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right')
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Logistic Regression Model Results')
plt.show()    
```

```python
#confusion matrix for data tree
conmat = confusion_matrix(csplit[3], dt_predictions)

val = np.mat(conmat) 
classnames = list(set(csplit[2]))
df_cm = pd.DataFrame(
        val, index=classnames, columns=classnames, 
    )

plt.figure()
heatmap = sns.heatmap(df_cm, annot=True, cmap="Purples")
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right')
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Decision Tree Model Results')
plt.show()    
```

I did run into some errors trying to perform regression when I forgot to convert the strings into numerical values, but other than that doing the models was pretty straightforward.

The hardest part I encountered was with creating the charts and trying to figure out how to visualize the regression models; since they have many features, a traditional plot wouldn't work. After doing some research, I decided to make multiple regression plots, one for each variable, so that we could see roughly how the relationships the models will learn are. For instance, we can see that region doesn't really affect the cost, but age does. Smokers causes charges to increase, as do males, while non-smokers and being female drives charges down.

For charting the classifier models, I was unsure of how to approach them visually as well. I chose to make confusion matrixes, since they would illustrate the spread of predictions and true values in an easy to understand fashion. I also plotted the decision tree itself.

The biggest understanding I gained is how complex relationships can be between data. It's easy to underestimate it until you try to figure out how to chart relationships yourself and realize you have multiple features all trying to be related to the final label.
